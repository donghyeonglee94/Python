# 환경 설정
- [Python 3.8.10](https://www.python.org/downloads/release/python-3810/) 설치
- [Visual Studio Code](https://code.visualstudio.com/) 설치

### Python3.8.10 설치
1. Python 사이트 접속
2. Downloads 탭으로 이동
3. Python 3.8.10 클릭
4. OS에 맞는 설치 파일 다운로드

5. 설치 파일 실행
6. 설치시 all user 체크
7. 설치 완료후 PowerShell을 실행
8. cmd 입력 후 python --version으로 설치 확인

### Visual Studio Code 설치
1. Visual Studio 접속
2. Download
3. 다운로드 후 실행 확인

### Python 가상환경 실행
1. PowerShell 실행
2. cmd 명령어 입력
3. (python -m venv 폴더명)으로 가상환경 폴더 생성
4. (cd 가상환경폴더)로 이동
5. OS별 가상환경 활성
- Mac : source bin\activate
- Windows : Scripts\activate
