# Exception
- 예외 발생시 except 코드 실행
- raise: 예외 발생
- assert: 지정된 조건식이 거짓일 때 AssertionError 예외 발생, 디버깅 모드에서만 실행

(+) 파이썬은 기본적으로 디버깅 모드(__debug__ = True), python -O 옵션으로 False로 실행

```python
try:
    실행 코드
    raise Exception('메시지')
except 예외 as e:
    예외 발생시 실행 코드
else:
    예외 없는 경우 실행 코드
finally:
    항상 실행되는 코드
    
assert 조건식, '에러 메시지'
```

[예외 계층도](https://docs.python.org/3/library/exceptions.html)


````python
try:
    for i in range(1, 10):
        if (i == 5):
            raise Exception("5 예외")
        print(i)
except Exception as e:
    print(e)
finally:
    print("프로그램이 종료됩니다!!")
````

### 예외 만들기

```python
class 예외명(Exception):
    def __init__(self):
        super().__init__('에러메시지')
```

````python
class NumberError(Exception):
    def __init__(self):
        super().__init__('5의 배수가 아닙니다.')
    
    
def typing_print():
    try:
        x = int(input('숫자를 입력하세요: '))
        if x % 5 != 0:                     # x가 3의 배수가 아니면
            raise NumberError    # NotThreeMultipleError 예외를 발생시킴
        print(x)
    except Exception as e:
        print('예외가 발생했습니다.', e)

typing_print()
````
